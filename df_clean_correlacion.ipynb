```python
data = './Data/Imputed_DF_DTR.csv'

df = pd.read_csv(data)
```

```python
import pandas as pd

# Supongamos que 'df' es tu DataFrame y ya contiene las variables de interés

# Lista de columnas para verificar outliers
columns_to_check = ['CICLO', 'FACTOREL', 'AOI', 'DISP', 'ITBU', 'TRAREM', 'CURSNR', 'CURSR',
       'NFORMA', 'NAC1', 'ECIV1', 'EDADN', 'CCAA', 'NPADRE', 'NMADRE', 'PROV',
       'NCONY', 'SEXO1', 'NVIVI', 'RELPP1', 'NIVEL', 'NPERS', 'OFEMP',
       'EDADEST', 'SIDAC1', 'SIDAC2', 'FOBACT', 'ASALA', 'MUN1', 'EMBUS',
       'PRONA1', 'AUSENT', 'AYUDFA', 'EMPANT', 'BUSCA', 'SIDI1', 'NUEVEM',
       'DTANT']
```

```python
# Diccionario para almacenar los outliers
outliers = {}

for column in columns_to_check:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 5 * IQR
    upper_bound = Q3 + 5 * IQR
    
    # Identificar outliers
    mask = (df[column] < lower_bound) | (df[column] > upper_bound)
    outliers[column] = df[mask]
    
    # Imprimir los outliers encontrados para cada variable
    print(f"Outliers in {column}:")
    print(df.loc[mask, column], "\n")  # Imprime solo los valores outliers para la columna actual

# 'outliers' contiene los DataFrames de outliers para cada columna revisada
```

```python
import pandas as pd
from sklearn.impute import KNNImputer

# Cargando los datos
# data = './EPA_2005_2023_FULL.csv'
# df = pd.read_csv(data)

# Mantener solo las columnas numéricas para calcular correlaciones
df_numeric = df_clean.select_dtypes(include=[np.number])

# Calcular correlaciones
correlaciones = df_numeric.corr()

# Seleccionar correlaciones para 'ITBU', omitiendo su propia correlación
df_correlaciones = correlaciones['ITBU'].drop('ITBU')

# Obtener los nombres de las 10 características más correlacionadas en valor absoluto
definitivas = df_correlaciones.abs().nlargest(10).index.tolist()

# Añadir 'ITBU' a la lista de características definitivas
definitivas.append('ITBU')

# Filtrar el dataframe original para mantener solo las columnas seleccionadas
df_final = df[definitivas]

# Aplicar KNNImputer al dataframe filtrado
imputer = KNNImputer(n_neighbors=5)
df_imputed = pd.DataFrame(imputer.fit_transform(df_final), columns=df_final.columns)
```
